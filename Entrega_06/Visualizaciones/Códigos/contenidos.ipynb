```python
import pandas as pd
df = pd.read_csv("Base_marvelstarwars.csv", sep=";", encoding="utf-8")


df['AÑO'] = pd.to_numeric(df['AÑO'], errors='coerce')

df = df.dropna(subset=['AÑO', 'CONTENIDO'])
```

```python
import altair as alt
import pandas as pd

df['AÑO'] = df['AÑO'].astype(int)

conteo_contenido = (
    df.groupby(['AÑO', 'CONTENIDO'])
      .size()
      .reset_index(name='cantidad')
)

alt.Chart(conteo_contenido).mark_line(point=True).encode(
    x=alt.X('AÑO:Q', sort='ascending', title='Año'),
    y=alt.Y('cantidad:Q', title='Número de películas'),
    color=alt.Color('CONTENIDO:N', title='Tipo de contenido'),
    tooltip=['AÑO', 'CONTENIDO', 'cantidad']
).properties(
    width=700,
    height=400,
    title='Número de películas por tipo de contenido'
)
```
