```python
import pandas as pd
import altair as alt

df = pd.read_csv("Base_marvel_starwars.csv", sep=";")


df["CONTENIDO"] = df["CONTENIDO"].str.strip().str.lower()

df["categoria"] = df["CONTENIDO"].apply(
    lambda x: "Original" if x == "original" else "Historia existente"
)

df["AÑO"] = pd.to_numeric(df["AÑO"], errors="coerce")

df_grouped = df.groupby(["AÑO", "categoria"]).size().reset_index(name="cantidad")

chart = alt.Chart(df_grouped).mark_area().encode(
    x=alt.X("AÑO:O", title="Año"),   
    y=alt.Y("cantidad:Q", stack="zero", title="Cantidad de películas"),
    color=alt.Color("categoria:N", title="Categoría"),
    tooltip=["AÑO", "categoria", "cantidad"]
).properties(
    width=800,
    height=400,
    title="Comparación de películas: Original vs Historia existente" 
)

chart
```
